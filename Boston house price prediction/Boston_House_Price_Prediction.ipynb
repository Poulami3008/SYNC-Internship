```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
data=pd.read_csv('HousingData.csv')
```

```python
data[data.isnull().any(axis=1)]
```

```python
Null_per=round((((data.isnull().sum())/(len(data)))*100),2)
round(Null_per.sort_values(ascending=False),2)
```

```python
#replacing the null values with mean value of corresponding column
data['CRIM'].fillna(int(data['CRIM'].mean()), inplace=True)
data['ZN'].fillna(int(data['ZN'].mean()), inplace=True)
data['INDUS'].fillna(int(data['INDUS'].mean()), inplace=True)
data['CHAS'].fillna(int(data['CHAS'].mean()), inplace=True)
data['AGE'].fillna(int(data['AGE'].mean()), inplace=True)
data['LSTAT'].fillna(int(data['LSTAT'].mean()), inplace=True)
data.head()
```

```python
data.isnull().any()
```

### Visualizing the data

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), cmap='GnBu', annot=True)
plt.show()
```

```python
fig, ax=plt.subplots(ncols=4, nrows=4, figsize=(20, 20))
index=0
ax=ax.flatten()

for col, value in data.items():
    sns.boxplot(y=col, data=data, ax=ax[index])
    index+=1
plt.show()
```

```python
fig, ax=plt.subplots(ncols=4, nrows=4, figsize=(20, 20))
index=0
ax=ax.flatten()

for col, value in data.items():
    sns.distplot(value, ax=ax[index])
    index+=1
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
Model=Scaler.fit(data)
Scaled_data=Model.transform(data)

Scaled_data=pd.DataFrame(Scaled_data, columns=data.columns)
Scaled_data.head()
```

```python
fig, ax=plt.subplots(ncols=4, nrows=4, figsize=(20, 20))
index=0
ax=ax.flatten()

for col, value in Scaled_data.items():
    sns.distplot(value, ax=ax[index])
    index+=1
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

St_sc = StandardScaler()
St_sc_model= St_sc.fit(data)
St_sc_data=St_sc_model.transform(data)

St_sc_data=pd.DataFrame(St_sc_data, columns=data.columns)
St_sc_data.head()
```

```python
fig, ax=plt.subplots(ncols=4, nrows=4, figsize=(20, 20))
index=0
ax=ax.flatten()

for col, value in St_sc_data.items():
    sns.distplot(value, ax=ax[index])
    index+=1
plt.show()
```

```python
x=data.drop(columns='MEDV', axis=1)
y=data['MEDV']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.2, random_state = 0)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
train_data_prediction=dt.predict(x_train)
train_data_prediction
```

```python
from sklearn import metrics

print("r2_score:", metrics.r2_score(y_train, train_data_prediction))
print("MSE:", metrics.mean_squared_error(y_train, train_data_prediction))
print("MAE:", metrics.mean_absolute_error(y_train, train_data_prediction))
```

```python
plt.scatter(y_train, train_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual Vs predited')
plt.show()
```

```python
test_data_prediction=dt.predict(x_test)
test_data_prediction
```

```python
print("r2_score:", metrics.r2_score(y_train, train_data_prediction))
print("MSE:", metrics.mean_squared_error(y_train, train_data_prediction))
print("MAE:", metrics.mean_absolute_error(y_train, train_data_prediction))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
train_data_prediction=rf.predict(x_train)
train_data_prediction
```

```python
print("r2_score:", metrics.r2_score(y_train, train_data_prediction))
print("MSE:", metrics.mean_squared_error(y_train, train_data_prediction))
print("MAE:", metrics.mean_absolute_error(y_train, train_data_prediction))
```

```python
test_data_prediction=rf.predict(x_test)
test_data_prediction
```

```python
print("r2_score:", metrics.r2_score(y_test, test_data_prediction))
print("MSE:", metrics.mean_squared_error(y_test, test_data_prediction))
print("MAE:", metrics.mean_absolute_error(y_test, test_data_prediction))
```

```python
plt.scatter(y_train, train_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual Vs predited')
plt.show()
```
